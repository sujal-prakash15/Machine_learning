# Polynomial Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
from google.colab import drive
drive.mount('/content/drive')
```

```
df=pd.read_csv('/content/drive/MyDrive/Polynomial Regression/Position_Salaries.csv')
```

```
print(df)
```

```
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
```

```
print(x)
```

```
print(y)
```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x,y)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
X_poly=poly_reg.fit_transform(x)
reg_2=LinearRegression()
reg_2.fit(X_poly,y)
```

```
print(X_poly)
```

## Visualising the Linear Regression results

```
plt.scatter(x,y,color='red')
plt.plot(x,reg.predict(x),color='blue')
plt.title('Truth or bluff(Linear Regression')
plt.xlabel('postion level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results

```
plt.scatter(x,y,color='red')
plt.plot(x,reg_2.predict(X_poly),color='blue')
plt.title('Truth vs Bluff(Polynomial Reg)')
plt.xlabel('position level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```
X_grid=np.arange(min(x),max(x),0.1)
X_grid=X_grid.reshape(len(X_grid),1)
plt.scatter(x,y,color='red')
plt.plot(X_grid,reg_2.predict(poly_reg.fit_transform(X_grid)),color='blue')
plt.title('Truth vs Bluff(Polynomial Reg)')
plt.xlabel('position level')
plt.ylabel('Salary')
plt.show()
#over-fitted model but don't worry as we have to predict x=[0,10]
```

## Predicting a new result with Linear Regression

```
reg.predict([[6.5]])
```

## Predicting a new result with Polynomial Regression

```
reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
